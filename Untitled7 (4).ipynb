Importing DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df=pd.read_csv("loan_predict.zip")
```

CORRELATION HEATMAP

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

DROPPING IRRELEVANT COLUMN

```python
df = df.drop("LoanID", axis=1)
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().count()
```

ONE HOT ENCODING

```python
binary_cols = ["HasMortgage", "HasDependents", "HasCoSigner"]

for col in binary_cols:
    df[col] = df[col].map({"No": 0, "Yes": 1})
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),["Education","EmploymentType","MaritalStatus",	"LoanPurpose"	])],
                     remainder="passthrough")
```

```python
 df = pd.DataFrame(
    ct.fit_transform(df),
    columns=ct.get_feature_names_out()
)
```

```python
df.info()
```

```python
type(df)
```

```python
df.columns
```

TRAIN TEST AND SPLITTING DATASET

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=29)
```

```python
X_train
```

LOGISTIC REGRESSION

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
lr.predict(X_test)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_auc_score
```

```python
accuracylr=accuracy_score(y_test,y_pred)
print("accuracy_score:",accuracylr)
cm_lr=confusion_matrix(y_test,y_pred)
print("confusion matrix:\n",cm_lr)
print("classification report:\n ",classification_report(y_test,y_pred))
roc_lr = roc_auc_score(y_test, y_pred)
print("ROC-AUC:", roc_lr)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.show()
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred1=dt.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(
    dt,
    max_depth=2,
    filled=True,
    feature_names=X.columns,
    class_names=True
)
plt.show()
```

```python
accuracydt=accuracy_score(y_test,y_pred1)
print("accuracy_score:",accuracydt)
cm_dt=confusion_matrix(y_test,y_pred1)
print("confusion matrix:\n",cm_dt)
print("classification report:\n ",classification_report(y_test,y_pred1))
roc_dt = roc_auc_score(y_test, y_pred1)
print("ROC-AUC:", roc_dt)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred1)
plt.show()
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred2=rf.predict(X_test)
```

```python
accuracy_rf=accuracy_score(y_test,y_pred2)
print("accuracy_score:",accuracy_rf)
rf_dt=confusion_matrix(y_test,y_pred2)
print("confusion matrix:\n",rf_dt)
print("classification report:\n ",classification_report(y_test,y_pred2))
roc_rf = roc_auc_score(y_test, y_pred2)
print("ROC-AUC:", roc_rf)
```

```python
importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf.feature_importances_
})

importance = importance.sort_values(by='Importance', ascending=False)
print(importance)
```

```python
plt.barh(X_train.columns, rf.feature_importances_)
plt.xlabel("Importance")
plt.title("Feature Importance")
plt.show()
```

K NEAREST NEIGHBOURS

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred3=knn.predict(X_test)
```

```python
accuracy_knn=accuracy_score(y_test,y_pred3)
print("accuracy_score:",accuracy_knn)
cm_knn=confusion_matrix(y_test,y_pred3)
print("confusion matrix:\n",cm_knn)
print("classification report:\n ",classification_report(y_test,y_pred3))
roc_knn = roc_auc_score(y_test, y_pred3)
print("ROC-AUC:", roc_knn)
```

XGBOOST

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state=42)
xgb.fit(X_train, y_train)

y_pred4 = xgb.predict(X_test)
```

```python
accuracy_XGB=accuracy_score(y_test,y_pred4)
print("accuracy_score:",accuracy_XGB)
cm_XGB=confusion_matrix(y_test,y_pred4)
print("confusion matrix:\n",cm_XGB)
print("classification report:\n ",classification_report(y_test,y_pred4))
roc_XGB = roc_auc_score(y_test, y_pred4)
print("ROC-AUC:", roc_XGB)
```

```python
from xgboost import plot_importance
plot_importance(xgb)
plt.show()
```

```python
results = {
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "KNN",
        "XGBoost"
    ],
    "Accuracy": [
        accuracylr,
       accuracydt,
        accuracy_rf,
       accuracy_knn,
        accuracy_XGB
    ]
}

comparison = pd.DataFrame(results)
print(comparison)
```
